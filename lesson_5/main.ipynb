```python
import pandas as pd

df = pd.read_csv('data.csv')
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])
data = df.copy()
df['month_added'] = df['dateAdded'].apply(lambda x: x.month)
```

```python
data[
    data['name'] == 'Taco Bell'
]['city'].value_counts()
```

```python
data.groupby(data['dateAdded'].dt.month)['id'].nunique().to_frame()
```

```python
df.loc[df['name'] == 'Taco Bell', ['city', 'address']].groupby('city')['address'].nunique().sort_values(ascending=False)
```

```python
data['update_delta'] = (data['dateUpdated'] - data['dateAdded']).apply(lambda x: x.days)

result = data.groupby('city', as_index=False)[['update_delta', 'latitude']].agg({
    'update_delta': 'mean',
    'latitude': 'max',
})
zep_mean = result.loc[result['city'] == 'Zephyrhills', 'update_delta']
```
